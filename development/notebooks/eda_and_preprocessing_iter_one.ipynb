# Final preprocessing

```python
df_null_dropped = df_original.dropna(axis=1)
df_null_dropped
```

```python
columns_to_drop = ['Lineitem requires shipping', 'Lineitem taxable', 'Lineitem fulfillment status']

df_to_use = df_null_dropped.drop(columns=columns_to_drop)
df_to_use
```

# Normalising Name

```python
names = df_to_use['Name'].value_counts()

with pd.option_context('display.max_rows', None):
    print(names)
    
df_to_use['Name'].nunique()
```

```python
# we see there are some weird names
df_to_use[df_to_use['Name'].astype(str).str.len() > 6]
```

```python
# we also see some Names share the same number, just whether it starts with # or not

df_to_use.drop_duplicates(subset='Name', keep='first', inplace=True)
df_to_use['NumericPart'] = df_to_use['Name'].astype(str).str.replace('#', '')
df_to_use[df_to_use.duplicated(subset='NumericPart', keep=False)].sort_values(by='NumericPart')
```

Assuming they are the same person

```python
# Handling the weird names
filtered_rows = df_to_use[df_to_use['Name'].astype(str).str.len() > 6]
unique_names = filtered_rows['Name'].unique()
name_mapping = {name: idx + 10000000 for idx, name in enumerate(unique_names)}
df_to_use.loc[filtered_rows.index, 'Name'] = df_to_use.loc[filtered_rows.index, 'Name'].map(name_mapping)

# Merge the numbers with '#' from 'Name' column
df_to_use['Name'] = df_to_use['Name'].astype(str).str.replace('#', '')

# Get unique 'Name' values and assign unique IDs
name_to_id = {}
current_id = 0
for name in df_to_use['Name'].unique():
    name_to_id[name] = current_id
    current_id += 1

# Replace 'Name' values with corresponding IDs
df_to_use['Name'] = df_to_use['Name'].map(name_to_id)

# Rename 'Name' column to 'customer_id'
df_to_use = df_to_use.rename(columns={'Name': 'customer_id'})

df_to_use_names_fixed = df_to_use.copy()
```
